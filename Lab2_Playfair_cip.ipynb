```python
import string

def toLowerCase(text):
  return text.lower()

def removeSpaces(text):
  return "".join(text.split())

def Diagraph(text):
  return [text[i:i+2] for i in range(0, len(text), 2)]

def FillerLetter(text):
  for i in range(0, len(text) - 1, 2):
    if text[i] == text[i+1]:
      text = text[:i+1] + 'x' + text[i+1:]
  return text if len(text) % 2 == 0 else text + 'x' 

def generateKeyTable(word):
  key_letters = []
  for char in word.lower() + string.ascii_lowercase:
    if char not in key_letters and char != 'j':
      key_letters.append(char)
  return [key_letters[i:i+5] for i in range(0, 25, 5)]

def search(mat, element):
  for i, row in enumerate(mat):
    if element in row:
      return i, row.index(element)
  return None 

def encrypt_RowRule(matr, e1r, e1c, e2r, e2c):
  return matr[e1r][(e1c+1)%5], matr[e2r][(e2c+1)%5]

def encrypt_ColumnRule(matr, e1r, e1c, e2r, e2c):
  return matr[(e1r+1)%5][e1c], matr[(e2r+1)%5][e2c]

def encrypt_RectangleRule(matr, e1r, e1c, e2r, e2c):
  return matr[e1r][e2c], matr[e2r][e1c]

def encryptByPlayfairCipher(Matrix, plainlist):
  CipherText = []
  for pair in plainlist:
    pair = pair.replace('j', 'i') 
    pair = pair.replace('J', 'I')

    result1 = search(Matrix, pair[0])
    result2 = search(Matrix, pair[1])

    if result1 is None or result2 is None:
      raise ValueError(f"Character not found in matrix: {pair}")

    ele1_x, ele1_y = result1
    ele2_x, ele2_y = result2

    if ele1_x == ele2_x:
      c1, c2 = encrypt_RowRule(Matrix, ele1_x, ele1_y, ele2_x, ele2_y)
    elif ele1_y == ele2_y:
      c1, c2 = encrypt_ColumnRule(Matrix, ele1_x, ele1_y, ele2_x, ele2_y)
    else:
      c1, c2 = encrypt_RectangleRule(Matrix, ele1_x, ele1_y, ele2_x, ele2_y)

    CipherText.append(c1 + c2)

  return CipherText

def main():
  try:
    text_Plain = input("Enter your Plaintext: ")
    text_Plain = removeSpaces(toLowerCase(text_Plain))
    text_Plain = FillerLetter(text_Plain)
    PlainTextList = Diagraph(text_Plain)

    key = input("Enter the Key: ")
    print("Key text:", key)

    Matrix = generateKeyTable(key)
    print("Plain Text:", text_Plain)

    CipherList = encryptByPlayfairCipher(Matrix, PlainTextList)
    CipherText = "".join(CipherList)
    print("CipherText:", CipherText)

  except ValueError as e:
    print(f"Error: {e}")
  except Exception as e:
    print("An unexpected error occurred:", e)

if __name__ == "__main__":
  main()
```
